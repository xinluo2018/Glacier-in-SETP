```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path_result_glacier = 'data/ele-change-result/ele_change_glaciers.pkl'
```

```python
result_df = pd.read_pickle(path_result_glacier)
```

```python
ids = np.where((result_df['rate']>0) & (result_df['rate']<3))[0]
result_anomalous = result_df.iloc[ids, :]
print(result_anomalous.shape)
result_anomalous.head(2)
```

```python
gla_x = [float(geom.centroid.x) for geom in result_anomalous['geometry']]
gla_y = [float(geom.centroid.y) for geom in result_anomalous['geometry']]
```

```python
result_anomalous = result_anomalous[['rgi_id', 
                                    'rate', 
                                    'gla_area',                         
                                    ]]
result_anomalous['longitude'] = gla_x
result_anomalous['latitude'] = gla_y
result_anomalous = result_anomalous.round(3)
result_anomalous
```

```python
result_anomalous.to_csv('tmp.csv', index=False)
```
